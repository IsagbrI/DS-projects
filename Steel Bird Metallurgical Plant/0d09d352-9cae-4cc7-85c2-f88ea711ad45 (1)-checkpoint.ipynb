## Подготовка моделей ##

Разобьём даннные на выборки, проверим размер уобучающей выборки, так как он должен быть не менее 1700 строк. Затем переберём модели с различными гиперпараметрами и найдём лучшую.

```python
random_state  = 130125

X = table_c.drop(columns=['key', 'last_temperature'])
y = table_c['last_temperature']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=random_state)

print(f"Размер обучающей выборки: {X_train.shape[0]}")


scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

Строк достаточно, даннные разбиты. Приступим к работе с моделями.

```python
from sklearn.dummy import DummyRegressor

model = DummyRegressor(strategy='median')

model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)

print(f"MAE на тестовой выборке: {mae}")

r2 = r2_score(y_test, y_pred)
print(f"R2 на тестовой выборке: {r2}")

models_and_params = {
    #'LinearRegression': (
    #    LinearRegression(),
    #    {}
    #),
    #'RandomForestRegressor': (
    #    RandomForestRegressor(random_state=random_state),
    #    {'n_estimators': [100, 200],
    #     'max_depth': [10, 20, None]
    #    }
    #),
    'SVR': (
        SVR(),
        {'C': [1, 10, 100],
            'kernel': ['linear', 'rbf'],
            'gamma': ['scale', 'auto'],
            'epsilon': [0.1, 0.2],
            'degree': [2, 3]
       }
    #),
    #'GradientBoostingRegressor': (
    #    GradientBoostingRegressor(random_state=random_state),
    #    {'n_estimators': [100, 200], 
    #     'max_depth': [6, 10, None], 
    #     'learning_rate': [0.01, 0.1, 0.2]
    #    }
    )
}

best_model = None
best_score = float('-inf')
best_name = None

for model_name, (model, param_grid) in models_and_params.items():
    print(f"Training {model_name}...")
    grid = GridSearchCV(model, param_grid, scoring='neg_mean_absolute_error', cv=5, verbose=3)
    grid.fit(X_train, y_train)

    if grid.best_score_ > best_score:
        best_score = grid.best_score_
        best_model = grid.best_estimator_
        best_name = model_name

    print(f"Лучшая модель и ей гиперпараметры {model_name}: {grid.best_params_}")
    print(f"Лучшая MAE этой модели {model_name}: {grid.best_score_}")
```

```python
y_pred_test = best_model.predict(X_test)
mae_test = mean_absolute_error(y_test, y_pred_test)
print(f"MAE на тестовой выборке ({best_name}): {mae_test}")

r2_test = r2_score(y_test, y_pred_test)
print(f"R2 на тестовой выборке ({best_name}): {r2_test}")
```
